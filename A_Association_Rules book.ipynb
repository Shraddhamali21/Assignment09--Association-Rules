```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
book=pd.read_csv("E:/SHRADDHA/DA-DS Assignments/DS Assignments/A_Association Rules/book.csv")
book
```

```python
df=pd.get_dummies(book)
df.head()
```

# Case 1

MINIMUM SUPPORT: 0.1

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)
rules
```

```python
rules.sort_values('lift',ascending = True)[0:50]
```

```python
rules[rules.lift>2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(rules.support,rules.confidence)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support v Confidence')
```

```python
plt.scatter(rules.support,rules.lift)
plt.xlabel('Support')
plt.ylabel('Lift')
plt.title('Support v Lift')
```

```python
plt.scatter(rules.confidence,rules.lift)
plt.xlabel('Confidence')
plt.ylabel('Lift')
plt.title('Confidence v Lift')
```
